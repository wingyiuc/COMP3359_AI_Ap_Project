```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import seaborn as sns
sns.set_style("white");

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
def hit_rate(y_true, y_pred):
    return np.sum(y_true * y_pred > 0) / len(y_true)
```

```python
ARMA = pd.read_pickle('result_ARMA35_1day.pk')
ARMA.columns = ['Real','ARMA']

LSTM1 = pd.read_pickle('result_LSTM1.pk')
LSTM2 = pd.read_pickle('result_LSTM2.pk')
LSTM3 = pd.read_pickle('result_LSTM3.pk')
LSTM4 = pd.read_pickle('result_LSTM4.pk')
```

```python
result = ARMA.join(LSTM1['LSTM1']).join(LSTM2['LSTM2']).join(LSTM3['LSTM3']).join(LSTM4['LSTM4'])[25:]
```

## A sample of predictions

```python
fig, ax = plt.subplots(1,2,figsize=(15,4), sharex=True)

result[500:550][['Real','ARMA', 'LSTM4']].plot(ax = ax[0])
result[500:550][['LSTM1','LSTM2','LSTM3', 'LSTM4']].plot(ax = ax[1])
plt.show()
```

## ARMA Model Evaluation

```python
print("RMSE:     \t {:.3%}".format(np.sqrt(mean_squared_error(result['Real'], result['ARMA']))))
print("MAE:      \t {:.3%}".format(mean_absolute_error(result['Real'], result['ARMA'])))
print("Hit Rate: \t {:.3%}".format(hit_rate(result['Real'], result['ARMA'])))
```

## LSTM1 Model Evaluation

```python
print("RMSE:     \t {:.3%}".format(np.sqrt(mean_squared_error(result['Real'], result['LSTM1']))))
print("MAE:      \t {:.3%}".format(mean_absolute_error(result['Real'], result['LSTM1'])))
print("Hit Rate: \t {:.3%}".format(hit_rate(result['Real'], result['LSTM1'])))
```

## LSTM2 Model Evaluation

```python
print("RMSE:     \t {:.3%}".format(np.sqrt(mean_squared_error(result['Real'], result['LSTM2']))))
print("MAE:      \t {:.3%}".format(mean_absolute_error(result['Real'], result['LSTM2'])))
print("Hit Rate: \t {:.3%}".format(hit_rate(result['Real'], result['LSTM2'])))
```

## LSTM3 Model Evaluation

```python
print("RMSE:     \t {:.3%}".format(np.sqrt(mean_squared_error(result['Real'], result['LSTM3']))))
print("MAE:      \t {:.3%}".format(mean_absolute_error(result['Real'], result['LSTM3'])))
print("Hit Rate: \t {:.3%}".format(hit_rate(result['Real'], result['LSTM3'])))
```

## LSTM4 Model Evaluation

```python
print("RMSE:     \t {:.3%}".format(np.sqrt(mean_squared_error(result['Real'], result['LSTM4']))))
print("MAE:      \t {:.3%}".format(mean_absolute_error(result['Real'], result['LSTM4'])))
print("Hit Rate: \t {:.3%}".format(hit_rate(result['Real'], result['LSTM4'])))
```
