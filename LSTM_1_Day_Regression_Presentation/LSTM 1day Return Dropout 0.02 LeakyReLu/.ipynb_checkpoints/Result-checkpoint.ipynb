```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import seaborn as sns

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
def hit_rate(y_true, y_pred):
    return np.sum(y_true * y_pred > 0) / len(y_true)
```

```python
ARMA = pd.read_pickle('result_ARMA35_1day.pk')
ARMA.columns = ['Real','ARMA']

LSTM1 = pd.read_pickle('result_LSTM1.pk')
LSTM2 = pd.read_pickle('result_LSTM2.pk')
LSTM3 = pd.read_pickle('result_LSTM3.pk')
LSTM4 = pd.read_pickle('result_LSTM4.pk')
```

```python
result = ARMA.join(LSTM1['LSTM1']).join(LSTM2['LSTM2']).join(LSTM3['LSTM3']).join(LSTM4['LSTM4'])[25:]
# result = ARMA.join(LSTM1['LSTM1']).join(LSTM2['LSTM2']).join(LSTM3['LSTM3'])[25:]
```

```python
result[650:700].plot()
```

```python
print("{:.3%}".format(np.sqrt(mean_squared_error(result['Real'], result['ARMA']))))
print("{:.3%}".format(mean_absolute_error(result['Real'], result['ARMA'])))
print("{:.3%}".format(hit_rate(result['Real'], result['ARMA'])))
```

```python
print("{:.3%}".format(np.sqrt(mean_squared_error(result['Real'], result['LSTM1']))))
print("{:.3%}".format(mean_absolute_error(result['Real'], result['LSTM1'])))
print("{:.3%}".format(hit_rate(result['Real'], result['LSTM1'])))
```
